```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
```

```python
np.random.seed(42)
x = np.random.random(6)*100 + 50
y = 10000*(x + 100*np.random.random(6)) + 1e6

x = np.matrix(x).T
y = np.matrix(y).T

fig, ax = plt.subplots()
ax.scatter(x, y)

lin_model = LinearRegression()
lin_model.fit(x, y)

t = np.matrix(np.linspace(40, 160, 500)).T
pred = lin_model.predict(t)
ax.axis([30,170,1500000,3600000])
ax.plot(t, pred)
plt.xlabel("Kvadratmeter")
plt.ylabel("Pris")

plt.show()
```
