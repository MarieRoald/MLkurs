```python
x = np.array([[1.5,1.85,2.6,2.97,3.1,3.5]]).T
y = np.array([[1.2,2.1,2.5,3.1,3.4,4.1]]).T

reg = LinearRegression()
reg.fit(x,y)

t = np.linspace(1,4,1001)
l = reg.predict(t[:,np.newaxis])

figure(figsize=(12,4))
subplot(1,3,1)
plot(t,l)
plot(x,y,'ro')
title('A')

axis([1,4,0,5])

from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=10)
x_p =  poly.fit_transform(x)

reg_p2 = LinearRegression()
reg_p2.fit(x_p,y)


t = np.linspace(1,4,1001)
l = reg_p2.predict(poly.transform(t[:,np.newaxis]))
subplot(1,3,2)
plot(t,l)
plot(x,y,'ro')
title('B')

subplot(1,3,3)

y2 = x**6/1000.

reg.fit(x,y2)


l = reg.predict(t[:,np.newaxis])

plot(x,y2,'ro')
plot(t,l)
title('C')

show()
```
