```python
from sklearn.datasets import load_iris
from sklearn import tree
import numpy as np

from sklearn.svm import SVC, LinearSVC
from sklearn.neighbors import KNeighborsClassifier

from sklearn.tree import DecisionTreeClassifier

from matplotlib.pyplot import *
import seaborn
from pandas import *
```

```python
from sklearn.datasets import make_classification
from useful_tools import plot_boundary

X1, Y1 = make_classification(n_features=2, n_redundant=0, n_informative=2,
                             n_clusters_per_class=1,flip_y=0.0,random_state=97)
```

```python
figure()

clf = DecisionTreeClassifier()
clf.fit(X1,Y1)

plot_boundary(X1,clf,padding=0.1,plot_step=0.005)
scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1)
show()
```

```python
figure()

from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
clf.fit(X1,Y1)

plot_boundary(X1,clf,padding=0.1,plot_step=0.005)
scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1)
show()
```

```python
figure()


clf = KNeighborsClassifier()
clf.fit(X1,Y1)

plot_boundary(X1,clf,padding=0.1,plot_step=0.005)
scatter(X1[:, 0], X1[:, 1], marker='o', c=Y1)
show()
```

```python
x = np.array([[1.5,1.85,2.6,2.97]]).T
y = np.array([[1.2,2.45,2.5,3.1]]).T

from sklearn.linear_model import LinearRegression
reg = LinearRegression()

reg.fit(x,y)

t = np.linspace(1,3.5,1001)
l = reg.predict(t[:,np.newaxis])

figure(figsize=(12,4))
subplot(1,3,1)
plot(t,l)
plot(x,y,'ro')
axis([1,3.5,1,4])
title('A')




x2 = np.array([[1.5,2.6,2.97]]).T
y2 = np.array([[1.2,2.5,3.1]]).T
reg.fit(x2,y2)

t = np.linspace(1,3.5,1001)
l = reg.predict(t[:,np.newaxis])


subplot(1,3,2)
plot(t,l)
plot(x,y,'ro')

axis([1,3.5,1,4])
title('B')

x3 = np.array([[1.85,2.97]]).T
y3 = np.array([[2.45,3.1]]).T

reg.fit(x3,y3)

t = np.linspace(1,3.5,1001)
l = reg.predict(t[:,np.newaxis])


subplot(1,3,3)
plot(t,l)
plot(x,y,'ro')

axis([1,3.5,1,4])
title('C')



show()
```

```python
x = np.array([[1.5,1.85,2.6,2.97,3.1,3.5]]).T
y = np.array([[1.2,2.1,2.5,3.1,3.4,4.1]]).T

reg = LinearRegression()
reg.fit(x,y)

t = np.linspace(1,4,1001)
l = reg.predict(t[:,np.newaxis])

figure(figsize=(12,4))
subplot(1,3,1)
plot(t,l)
plot(x,y,'ro')
title('A')

axis([1,4,0,5])

from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=10)
x_p =  poly.fit_transform(x)

reg_p2 = LinearRegression()
reg_p2.fit(x_p,y)


t = np.linspace(1,4,1001)
l = reg_p2.predict(poly.transform(t[:,np.newaxis]))
subplot(1,3,2)
plot(t,l)
plot(x,y,'ro')
title('B')

subplot(1,3,3)

y2 = x**6/1000.

reg.fit(x,y2)


l = reg.predict(t[:,np.newaxis])

plot(x,y2,'ro')
plot(t,l)
title('C')

show()
```
