```python
x = np.array([[1.5,1.35,2.3,2.97,3.1,4]]).T
y = np.array([[0.9,2.45,1.8,3.1,4.2,4]]).T

klasse = np.array([[0,0,0,1,1,1]]).T

reg = LinearRegression()

reg.fit(x,y)

pad = 0.5
x_max = np.max(x)
x_min = np.min(x)
y_min, y_max = np.min(y), np.max(y)


t = np.linspace(x_min-pad,x_max+pad,1001)
l = reg.predict(t[:,np.newaxis])

figure(figsize=(12,4))
subplot(1,3,1)
plot(t,l)
plot(x,y,'ro')
axis([x_min-pad,x_max+pad,y_min-pad,y_max + pad])
title('A')

## plot 2

from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=3)
x_p =  poly.fit_transform(x)

reg_p2 = LinearRegression()
reg_p2.fit(x_p,y)

l_p2 = reg_p2.predict(poly.fit_transform(t[:,np.newaxis]))


subplot(1,3,2)
plot(t,l_p2)
plot(x,y,'ro')

axis([x_min-pad,x_max+pad,y_min-pad,y_max + pad])
title('B')

# plot 3
tree = DecisionTreeClassifier()
X = np.concatenate([x,y],1)
tree.fit(X,klasse)

subplot(1,3,3)

plot_boundary(X,tree,padding=pad)
scatter(x,y,c=klasse)



axis([x_min-pad,x_max+pad,y_min-pad,y_max + pad])
title('C')

show()
```
